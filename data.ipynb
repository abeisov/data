```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['longitude'], df['latitude'], c=df['mag'], cmap='Blues', alpha=1, edgecolors='red')
plt.colorbar(label='Mag')
plt.title('Place')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', hue='mag', size='mag', sizes=(20, 200),data=df, palette='PuBuGn', alpha=0.7)
plt.title('Place ')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Mag')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['depth'], df['mag'], alpha=0.5, c=df['mag'], cmap='viridis')
plt.colorbar(scatter, label='Magnitude')
plt.title('Magnitude Depth')
plt.xlabel('Depth')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
plt.hist(df['mag'], bins=30, color='blue', edgecolor='black')
plt.title('Histogram of mag')
plt.xlabel('Mag')
plt.ylabel('Freq')
plt.grid(True)
plt.show()

plt.figure(figsize=(15, 6))
plt.hist(df['depth'], bins=30, color='red', edgecolor='black')
plt.title('Histogram of depth')
plt.xlabel('Depth')
plt.ylabel('Freq')
plt.grid(True)
plt.show()
```
